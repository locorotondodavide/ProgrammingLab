### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
#1. calcolo delle potenze
numeri = [1, 2, 3, 4, 5]
quadrati = []

for n in numeri:
    quadrati.append(n**2)

print(quadrati) # Output: [1, 4, 9, 16, 25]

#con list comprehension
quadrati = [n**2 for n in numeri]

#2. modifica di stringhe
parole = ["python", "codice", "lista"]
maiuscole = []

for p in parole:
    maiuscole.append(p.capitalize())

print(maiuscole) 

#con list comprehension
maiuscole = [p.capitalize() for p in parole]

#3. filtraggio di numeri
punteggi = [12, 55, 34, 80, 22, 95]
voti_alti = []

for v in punteggi:
    if v > 50:
        voti_alti.append(v)

print(voti_alti) # Output: [55, 80, 95]

#con list comprehension
voti_alti = [v for v in punteggi if v > 50]
```

### **2. Vettore di Numeri Primi**

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**  
   _(Puoi scriverli direttamente nell'array)._

2. **Conta quanti numeri ci sono nel vettore** utilizzando la funzione `len()`.  **Ottieni lo stesso numero** accedendo all'attributo `.size` del vettore.

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
import numpy as np

#1: Creazione del vettore
primi = np.array([2, 3, 5, 7])
print("Vettore dei numeri primi:", primi)

#2: Conteggio degli elementi
conteggio_len = len(primi)
conteggio_size = primi.size

print(f"Conteggio con len(): {conteggio_len}")
print(f"Conteggio con .size: {conteggio_size}")

#3: Tipo di dato
#il tipo sarà intero perché non ci sono decimali.
tipo_dato = primi.dtype
print(f"Il tipo di dato rilevato è: {tipo_dato}")

#4: List Comprehension 
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

lista_primi_comp = [n for n in range(11) if is_prime(n)]
vettore_primi_auto = np.array(lista_primi_comp)

print("Vettore generato con list comprehension:", vettore_primi_auto)
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

```python
import numpy as np

#array con numeri da 1 a 5
a = np.arange(1, 6) 

#array estraendo i primi 3 elementi di a
b = a[:3]

#array c che è il reverse di a
c = a[::-1]

divisione_array = a / c

print("Array a:", a)
print("Array b (slicing):", b)
print("Array c (reverse):", c)
print("Divisione dei due array:", divisione_array)



# Trasformo gli array in liste
lista_a = list(a)
lista_c = list(c)

#list comprehension
divisione_lista = [x / y for x, y in zip(lista_a, lista_c)]

print("Lista a:", lista_a)
print("Lista c:", lista_c)
print("Divisione delle due liste:", divisione_lista)
```
